```python
dataset_path = "datasets/aug_il_dataset"

yaml_content = f"""
path: {dataset_path.split('/')[-1]}
train: images/train
val: images/val
nc: 6  # Number of classes
names: ['aegypti', 'albopictus', 'anopheles', 'culex', 'culiseta', 'japonicus/koreicus']
"""

with open(f"{dataset_path}/dataset.yaml", "w") as f:
    f.write(yaml_content)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
# Train the model
epochs, imgsz, batch, rect = 20, 640, 8, False
name = f"{imgsz}sz_{epochs}ep_{batch}b_r" if rect else f"{imgsz}sz_{epochs}ep_{batch}b"
results = model.train(
    data=f"{dataset_path}/dataset.yaml",  # Path to the dataset YAML
    epochs=epochs,  # Number of epochs
    imgsz=imgsz,  # Image size
    batch=batch,   # Batch size
    rect=rect,
    name=name, # Experiment name
    save_period=10,
    plots=True
)
```
